```python
# imports
import pandas as pd

import plotly.graph_objects as go
import plotly.express as px
```

```python
# load data
flights = pd.read_csv(
    "data/flights_train.csv",
    parse_dates=["flight_date"]
)

flights.sort_values(
    by=["flight_date"],
    inplace=True
)
```

```python
counts_dates = flights["flight_date"].value_counts()

fig = px.bar(
    x=counts_dates.index,
    y=counts_dates.values
)

fig.show()
```

```python
date = counts_dates.index[2]

flights.query("flight_date == @date").groupby(by=["from", "to"]).count()
```
