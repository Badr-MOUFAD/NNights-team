```python
# imports
import pandas as pd
import plotly.graph_objects as go

from nnights.enrich_jobs import (is_holiday, distance_next_holiday, 
                                 distance_previous_holiday)


# generate dates
date_range = pd.date_range(
    start="2021-01-01",
    end="2025-01-01",
    freq="D"
)

# put in df
df_dates = pd.DataFrame(date_range, columns=["date"])

# compute distance next and distance previous to holiday
df_dates["distance_next_holiday"] = df_dates["date"].apply(lambda d: distance_next_holiday(d))
df_dates["distance_previous_holiday"] = df_dates["date"].apply(lambda d: distance_previous_holiday(d))
df_dates["is_holiday"] = df_dates["date"].apply(lambda d: is_holiday(d))
```

```python
fig = go.Figure(
    data=[
        go.Scatter(
            x=df_dates["date"],
            y=df_dates["distance_previous_holiday"],
            marker_color=df_dates["is_holiday"].map({True: "red", False: "gray"}),
            mode="lines+markers"
        )
    ]
)

fig.update_layout(
    title="Distance to next holiday",
    yaxis_title="number of days"
)

fig.show()
```
