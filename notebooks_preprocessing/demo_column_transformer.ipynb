```python
# imports
import numpy as np
import pandas as pd

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# load data
flights = pd.read_csv(
    "data/flights_train.csv",
    parse_dates=["flight_date"]
)

flights.sort_values(
    by=["flight_date"],
    inplace=True
)
```

```python
flights.columns
```

```python
# init preprocessor
categorical_features = ["from", "to"]
categoric_transformer = OneHotEncoder()

numerical_features = ['avg_weeks', 'std_weeks']
numeric_transformer = StandardScaler()


preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categoric_transformer, categorical_features),
    ]
)
```

```python
# apply preprocessor
data = flights.drop(labels=["target", "flight_date"], axis=1)
```

```python
preprocessor.fit(data)
```

```python
data
```

```python
d = preprocessor.transform(data)
```

```python
pd.DataFrame(d.todense(), columns=preprocessor.get_feature_names_out())
```

```python
preprocessor.get_feature_names_out()
```
